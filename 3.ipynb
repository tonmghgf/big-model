```python
import re

# ===================== 原始文本定义 =====================
text = "Login: admin_01, IP: 192.168.1.1, Time: 10:30. Order: laptop_789, Price: $1200. Error: code_404 at 10:30. Note: admin_01 updated the system."
test_str_d = "Price: $1200. Error: code_404"

# ===================== 目标 A：提取用户名 =====================
pattern_a = r"^Login: (\w+)"
result_a = re.search(pattern_a, text)
username = result_a.group(1) if result_a else "未匹配到"

# ===================== 目标 B：提取价格数字（无$） =====================
pattern_b = r"(?<=\$)\d+"
result_b = re.findall(pattern_b, text)

# ===================== 目标 C：提取重复时间 =====================
pattern_c = r"(\d{2}:\d{2}).*\1"
result_c = re.search(pattern_c, text)
repeat_time = result_c.group(1) if result_c else "无重复"

# ===================== 目标 D：贪婪 vs 懒惰模式 =====================
# 贪婪模式
greedy_pattern = r":(.*):"
greedy_result = re.search(greedy_pattern, test_str_d).group(1)
# 懒惰模式
lazy_pattern = r":(.*?):"
lazy_result = re.search(lazy_pattern, test_str_d).group(1)

# ===================== 进阶任务：密码强度验证 =====================
def check_password(password):
    pattern = r"^(?=.*\d)(?=.*[a-z])(?!.*\s).{8,16}$"
    return "✅ 密码符合要求" if re.fullmatch(pattern, password) else "❌ 密码不符合安全策略"

# ===================== 统一输出结果 =====================
print("===== 正则表达式综合练习结果 =====")
print(f"目标A - 提取用户名：{username}")
print(f"目标B - 提取价格数字：{result_b}")
print(f"目标C - 重复时间点：{repeat_time}")
print(f"目标D - 贪婪模式匹配结果：\"{greedy_result}\"")
print(f"目标D - 懒惰模式匹配结果：\"{lazy_result}\"")

print("\n===== 密码强度验证器测试 =====")
print(f"密码 1234abcd：{check_password('1234abcd')}")
print(f"密码 12345678：{check_password('12345678')}")
print(f"密码 abc 123xyz：{check_password('abc 123xyz')}")
print(f"密码 123ab：{check_password('123ab')}")
```
